```python
import numpy as np
import pandas as pd
import matplotlib
import seaborn
```

```python
survey_df = pd.read_csv("Colony_loss.csv")

print (survey_df.head(0))
```

```python
results = pd.DataFrame(survey_df)
results
```

```python

loss = 0
no_loss = 0
march_alive = 0
sept_alive = 0
treat = 0
no_treat = 0
treat_noloss = 0
no_treat_noloss = 0
for i in range(len(results)):
    if (results['Sept Alive'][i]>results['March Alive'][i]):
        loss += 1
        if (str(results['Treat'][i])=="Yes"):
            treat += 1
        else:
            no_treat +=1
            
    else: 
        no_loss += 1
        if (str(results['Treat'][i])=="Yes"):
            treat_noloss += 1
        else:
            no_treat_noloss +=1
    
    march_alive += results['March Alive'][i]
    sept_alive += results['Sept Alive'][i]

print ("sept_alive = ", sept_alive)
print ("march alive = ",march_alive)

#set up width and precision for formatting with f strings
w = 5 #width of field
p = 3 #total digits or "precision"

respondents = len(results)
loss_percent_by_beek = (respondents-no_loss)/respondents*100
total_loss = (((sept_alive - march_alive))/sept_alive)*100
loss_treat = (respondents-treat)/respondents*100
loss_no_treat = (respondents-no_treat)/respondents*100

print ("Total number of beeks reporting = ", respondents)

print (f"Percent of beekepers that reported a loss = {loss_percent_by_beek:{w}.{p}}")

print (f"Total colony loss percent reported = {total_loss:{w}.{p}}")

print (f"Percentage of beeks that treated who experienced losses = {loss_treat:{w}.{p}}")

print (f"Percentage of beeks that did not treat who experienced losses = {loss_no_treat:{w}.{p}}")

print (treat_noloss)
print (no_treat_noloss)
print (no_loss)
print (f"Percentage of beekeepers that did not treat who experienced no losses")
```
